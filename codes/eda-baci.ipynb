```python
import pandas as pd
import duckdb
```

```python
countries = pd.read_excel('../../../MRIO Processing/dicts/countries.xlsx', sheet_name='BACI')
select = countries.loc[countries['baci_name'] == 'Kazakhstan', 'baci_num'].iloc[0]

products = pd.read_excel('../../../MRIO Processing/dicts/sectors.xlsx', sheet_name='hs02')

input = '../../../../Repos/baci/final/BACI_HS02_V202301.parquet'
```

```python
def top_importers(country, year, product=None):
    
    duckdb.sql(
        f"""
        SELECT t, i, j, baci_name_short, v

        FROM (
            SELECT t, i, j, sum(v) AS v
            FROM '{input}' 
            WHERE i={country} AND t={year} AND k={productcode}
            GROUP BY t, i, j
            ORDER BY v DESC
        ) AS t1
        
        JOIN (SELECT * FROM countries) AS t2
        ON t1.j = t2.	baci_num

        LIMIT 10
        """
    ).df()
```

### Top 10 exports in 2021

```python
duckdb.sql(
    f"""
    SELECT t, i, v, hs02, section_name, hs02_desc

    FROM (
        SELECT t, i, k, sum(v) AS v
        FROM '{input}' 
        WHERE i={select} AND t=2021
        GROUP BY t, i, k
        ORDER BY v DESC
    ) AS t1
    
    JOIN (SELECT * FROM products) AS t2
    ON t1.k = t2.hs02 

    LIMIT 10
    """
).df()
```
